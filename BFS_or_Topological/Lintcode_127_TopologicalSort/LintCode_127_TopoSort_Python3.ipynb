## Classical BFS Topological Sort Template

```python
"""
Definition for a Directed graph node
class DirectedGraphNode:
    def __init__(self, x):
        self.label = x
        self.neighbors = []
"""


class Solution:
    """
    @param: graph: A list of Directed graph node
    @return: Any topological order for the given graph.
    """
    def topSort(self, graph):
        if not graph:
            return
        
        indegree = {tail: 0 for tail in graph}
        
        for tail in graph:
            for head in tail.neighbors:
                indegree[head] += 1
        
        queue = []
        for node in indegree:
            if indegree[node] == 0:
                queue.append(node)
        
        order = []
        
        while queue:
            tail = queue.pop(0)
            order.append(tail)
            
            for head in tail.neighbors:
                indegree[head] -= 1
                if indegree[head] == 0:
                    queue.append(head)
        
        return order
```

## Classical Recursive DFS, no care about cycle

```python
"""
Definition for a Directed graph node
class DirectedGraphNode:
    def __init__(self, x):
        self.label = x
        self.neighbors = []
"""


class Solution:
    """
    @param: graph: A list of Directed graph node
    @return: Any topological order for the given graph.
    """
    def topSort(self, graph):
        # write your code here
        if not graph:
            return []
        
        visited = set()
        order = []
        for v in graph:
            if v not in visited:
                self.dfs(v, graph, visited, order)
        
        return order[::-1]
        
    def dfs(self, v, graph, visited, order):
        visited.add(v)
        for head in v.neighbors:
            if head not in visited:
                self.dfs(head, graph, visited, order)
        order.append(v)
        
```

## Classical Recursive DFS, does care about cycle

```python
"""
Definition for a Directed graph node
class DirectedGraphNode:
    def __init__(self, x):
        self.label = x
        self.neighbors = []
"""


class Solution:
    """
    @param: graph: A list of Directed graph node
    @return: Any topological order for the given graph.
    """
    def topSort(self, graph):
        # write your code here
        if not graph:
            return
        
        visited = {node: 0 for node in graph}
        
        order = []
        for node in graph:
            if visited[node] == 0 and not self.acyclic(node, graph, visited, order):
                return []
        order.reverse()
        return order
        
    def acyclic(self, node, graph, visited, order):
        if visited[node] == -1:
            return False
        
        if visited[node] == 1:
            return True
            
        visited[node] = -1
        for head in node.neighbors:
            if not self.acyclic(head, graph, visited, order):
                return False
        
        order.append(node)
        visited[node] = 1
        return True
```
